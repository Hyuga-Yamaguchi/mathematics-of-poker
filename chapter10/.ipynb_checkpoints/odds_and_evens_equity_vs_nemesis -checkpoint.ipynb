```python
import numpy as np
import matplotlib.pyplot as plt

# Label
plt.xlabel('p (X plays 0 pennies) ')
plt.ylabel('EQ')

# Function
def f1(x):
    return 2 * x - 1

def f2(x):
    return 1 - 2 * x

#Data
step = 0.02
dataset = {
    data1: {
        start: 0.0,
        end: 0.5,
        step: step,
        func: f1
    },
    data2: {
        start: 0.5,
        end: 1.0
        step: step,
        func: f2
    }
}

# Plot graph
for key in dataset:
    data = dataset[key]
    domain = np.arange(data[start], data[end], data[step])
    function = data.func
    plt.plot(domain, function)
plt.show()
```
