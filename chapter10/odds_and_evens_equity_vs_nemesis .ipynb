```python
import numpy as np
import matplotlib.pyplot as plt

# Function
def f1(x):
    return 2 * x - 1

def f2(x):
    return 1 - 2 * x

#Data
step = 0.02
dataset = {
    'data1': {
        'start': 0.0,
        'end': 0.5,
        'func': f1
    },
    'data2': {
        'start': 0.5,
        'end': 1.0,
        'func': f2
    }
}


# Plot graph
for key in dataset:
    data = dataset[key]
    domain = np.linspace(data['start'], data['end'])
    function = data['func']
    plt.plot(domain, function(domain), color='blue')

# Label
plt.xlabel('p (X plays 0 pennies) ')
plt.ylabel('EQ')    
plt.show()
```
